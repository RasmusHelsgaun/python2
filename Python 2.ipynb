# 1. Create a bar plot that shows distribution of german and british citizens in terms of age.

```python
%matplotlib notebook
import kkdata
import matplotlib.pyplot as plt
import cit

germany = 5180
britain = 5170
countries = [germany, britain]
age_dist = cit.get_citizen_dist_age(countries)
print(age_dist)

plt.figure()
title = 'Distribution of German and British citizens by age in {}'.format(2015)
plt.title(title, fontsize=12)
plt.xlabel("Ages", fontsize=10)
plt.ylabel("Amount", fontsize=10)

x_germany = age_dist[5180].keys()
y_germany = age_dist[5180].values()
plt.bar(x_germany, y_germany, width=0.5, linewidth=0, align='center', color='red')

x_britain = age_dist[5170].keys()
y_britain = age_dist[5170].values()
plt.bar(x_britain, y_britain, width=0.5, linewidth=0, align='center', color='blue')
```

# 2. How many French are there compared to Germans and how are they distributed over various neighbourhoods?

```python
%matplotlib notebook
import kkdata
import matplotlib.pyplot as plt
import cit

germany = 5180
france = 5130
countries = [germany, france]

total = cit.get_citizen_dist_total(countries)
print("Total distribution:", total)

plt.figure()
title = 'Total distribution of French and German citizens by area in {}'.format(2015)
plt.title(title, fontsize=12)
plt.xlabel("Area", fontsize=10)
plt.ylabel("Amount", fontsize=10)

x_total = total.keys()
y_total = total.values()
plt.bar(x_total, y_total, width=10, align='center')
```

```python
%matplotlib notebook
import kkdata
import matplotlib.pyplot as plt
import cit

germany = 5180
france = 5130
countries = [germany, france]

dist = cit.get_citizen_dist_area(countries)
print("Distribution per area:", dist)

plt.figure()
title = 'Distribution of French and German citizens by area in {}'.format(2015)
plt.title(title, fontsize=12)
plt.xlabel("Area", fontsize=10)
plt.ylabel("Amount", fontsize=10)

x_germany = dist[germany].keys()
y_germany = dist[germany].values()
plt.bar(x_germany, y_germany, width=0.5, linewidth=0, align='center', color='red')

x_france = dist[france].keys()
y_france = dist[france].values()
plt.bar(x_france, y_france, width=0.5, linewidth=0, align='center', color='blue')
```

# 3. Create a pie chart showing the 5 major citizen groups of age between 20-65

```python
%matplotlib notebook
import kkdata
import matplotlib.pyplot as plt
import cit

top_5 = cit.get_top_5()
print("Top 5 citizen groups: ", top_5)

labels = []
sizes = []

for country in top_5:
    labels.append(str(country))
    sizes.append(top_5[country])
    
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
